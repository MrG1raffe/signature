### A notebook to visualize the convergence of series of exponentials in the unicity proof

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

```python
lam = 0.7
mu = 1
```

```python
trunc = 150
a = np.empty(trunc + 1)

for k in range(trunc + 1):
    k_arr = np.arange(k)
    a[k] = (mu / lam - k_arr).prod() / factorial(k)
```

```python
a.sum()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

x_grid = np.linspace(0, 2, 1000) + 0.0001

pows = np.arange(trunc + 1)[:, None]

for N in range(4, trunc + 1):
    ax[0].plot(x_grid, a[:N] @ ((x_grid - 1)**pows[:N]))
ax[0].plot(x_grid, x_grid**(mu / lam), "k--")
ax[0].grid("on")

t_grid = np.linspace(0, 10, 10000)
for N in range(4, trunc + 1, 10):
    ax[1].plot(t_grid, a[:N] @ ((np.exp(-lam * t_grid) - 1)**pows[:N]), label=f"$k = {N - 1}$")
ax[1].plot(t_grid, np.exp(-mu * t_grid), "k--", label="$e^{-\mu t}$")
ax[1].grid("on")
ax[1].legend()
ax[1].set_ylim([-0.1, 1.1])

plt.show
```
